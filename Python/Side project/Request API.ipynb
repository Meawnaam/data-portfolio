```python
import time

user_posts = []
for i in range(20) :
    url = f"https://jsonplaceholder.typicode.com/posts/{i+1}"
    resp = requests.get(url)

    if resp.status_code == 200 :
        json_data = resp.json()
        user_posts.append(
            (
                json_data["userId"],
                json_data["id"],
                json_data["title"],
                json_data["body"]
            )
        )

        time.sleep(1)

    else:
        user_posts.append("error")

```

```python
user_posts[0:3]
```

```python
import pandas as pd
df = pd.DataFrame(user_posts, columns=["userId" , "id", "title", "body"])

df.head()
```
